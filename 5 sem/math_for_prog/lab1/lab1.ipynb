Также известно, что все элементы матрицы A, расположенные вне главной диагонали, равны нулю за исключением двух элементов: a1,2 = 0,57 и a2,1 = 0,03. Найдите: 1) определитель матрицы A; 2) след матрицы, обратной к A.


```python
import mpmath as mp
mp.mp.dps = 50
n = 80000
a = mp.mpf('0.99')
last = mp.mpf('1.01')
d = (last - a) / (n - 1)
alpha = a / d
log_p = n * mp.log(d) + mp.loggamma(alpha + n) - mp.loggamma(alpha)
p = mp.exp(log_p)
det_M = a * (a + d) - mp.mpf('0.57')*mp.mpf('0.03')
a11_a22 = a * (a + d)
det_A = det_M * p / a11_a22
print(float(det_A))
psi_sum = mp.digamma(alpha + n) - mp.digamma(alpha)
sum_inv = psi_sum / d
inv_a11 = 1/a
inv_a22 = 1/(a + d)
sum_3n = sum_inv - inv_a11 - inv_a22
trace_2 = (a + a + d ) / det_M
trace_inv = trace_2 + sum_3n
print(float(trace_inv))
```

## 1.12. Матрица Адамара H – это квадратная матрица размера n × n, составленная из чисел 1 и −1, столбцы которой удовлетворяют соотношению $H_n · H^T_n = n · I_n$, где $I_n$ – это единичная матрица порядка n, $H_0 ≡1$.

а) Выпишите H1, H2, H4 в явном виде;

б) Покажите, что H4 = H2 ⊗H2;

в)∗Покажите, что $H_{2^k} = H2 ⊗ H_{2^{k-1}}$, где 2 ⩽k ∈N, а ⊗ означает произведение Кронекера.

г) Используя графические возможности, изобразите матрицу Адамара для n = 2^6, положительному элементу матрицы отвечает «оранжевый цвет», отрицательному – «синий цвет».

```python
def hadamard(k):
    if k == 0:
        return np.array([[1]])
    else:
        h = hadamard(k-1)
        return np.block([[h, h], [h, -h]])
H = hadamard(6)
cmap = colors.ListedColormap(['blue', 'orange'])
plt.imshow((H + 1) / 2, cmap=cmap, interpolation='none')
plt.show()
```

## 1.15. Найдите собственный ортонормированный базис и матрицу в этом базисе самосопряженного оператора ϕ, заданного в некотором ортонормированном базисе (e) матрицей

 $ A^{(e)}_\phi = \begin{pmatrix}
17 & -8 & 4 \\ 
 -8 & 17 & -4 \\ 
 4 & -4 & 11
\end{pmatrix} $

```python
A = Matrix([[17, -8, 4], [-8, 17, -4], [4, -4, 11]])

# Находим собственные значения и векторы
eigenvals = A.eigenvects(simplify=True)  # Убеждаемся, что результаты упрощены

# Собственный ортонормированный базис
eigenbasis = []
for val_info in eigenvals:  # Итерируемся по кортежам (значение, кратность, векторы)
    eigenvalue, multiplicity, eigenvectors = val_info  # Распаковываем кортеж
    for v in eigenvectors:
        # Нормализуем вектор (делим на евклидову норму)
        norm = sqrt(sum(x**2 for x in v)).evalf()  # Вычисляем норму
        normalized_v = [x / norm for x in v]  # Нормализованный вектор
        eigenbasis.append(normalized_v)

print("Ортонормированный базис:", eigenbasis)

# Матрица оператора в новом базисе
# Поскольку матрица самосопряженная, в базисе из собственных векторов она диагональная
diagonal_matrix = Matrix.diag(*[val for val, _, _ in eigenvals])  # Диагональная матрица с собственными значениями
print("Матрица в новом базисе:", diagonal_matrix)
```

## 1.16. Найдите ортогональное преобразование, приводящее квадратичную функцию к главным осям

 $f(x_1, x_2, x_3, x_4) = 3x_1^2 + 8x_1x_2 −3x_2^2 + 4x_3^2 −4x_3x_4 + x_4^2$.

```python
A_16 = Matrix([[3, 4, 0, 0], [4, -3, 0, 0], [0, 0, 4, -2], [0, 0, -2, 1]])

# Находим собственные значения и векторы
eigenvals_16 = A_16.eigenvects(simplify=True)

# Ортогональная матрица перехода (собственные векторы, нормализованные)
normalized_vectors = []
for _, _, eigenvectors in eigenvals_16:
    for v in eigenvectors:
        v_matrix = Matrix(v)  # Преобразуем в Matrix
        norm = sqrt(v_matrix.dot(v_matrix)).evalf()  # Евклидова норма
        normalized_v = v_matrix / norm  # Нормализация
        normalized_vectors.append(normalized_v)

P_16 = Matrix.hstack(*normalized_vectors)
print("Ортонормированная матрица перехода P:", P_16)

# Проверяем ортогональность (P^T * P = I)
print("Проверка ортогональности (P^T * P):", P_16.T * P_16)

# Матрица в главных осях
D_16 = P_16.T * A_16 * P_16
print("Диагональная матрица в главных осях:", D_16.simplify())
```

## 1.17. Определите тип кривой второго порядка, её каноническое уравнение и найдите каноническую систему координат, если кривая задана уравнением:

a) $ 5x^2 + 4xy + 8y^2 −32x −56y + 80 = 0; $

b) $ 5x^2 + 12xy −22x −12y −19 = 0. $ 

```python
import numpy as np
from sympy import Matrix, sqrt, symbols, solve

# 1.17. Определить тип кривой второго порядка, её каноническое уравнение и найти каноническую систему координат
print("1.17")

# Общие параметры для классификации
def conic_type(a, b, c):
    disc = b**2 - 4*a*c
    if disc < 0:
        return "Эллипс"
    elif disc == 0:
        return "Парабола"
    else:
        return "Гипербола"

# а) 5x^2 + 4xy + 8y^2 - 32x - 56y + 80 = 0
a_a, b_a, c_a = 5, 4, 8
type_a = conic_type(a_a, b_a, c_a)
print("Тип кривой а):", type_a)

A_17a = Matrix([[5, 2], [2, 8]])  # Симметричная матрица (b/2)
b_vec_a = Matrix([-32, -56])  # Линейные коэффициенты
f_a = 80  # Константа

# Собственные значения
eigenvals_17a = A_17a.eigenvects(simplify=True)
lambdas_a = [val for val, _, _ in eigenvals_17a]
print("Собственные значения для а):", lambdas_a)

# Нормализованные собственные векторы для матрицы поворота P
normalized_vectors_17a = []
for _, _, eigenvectors in eigenvals_17a:
    for v in eigenvectors:
        v_matrix = Matrix(v)
        norm = sqrt(v_matrix.dot(v_matrix))
        normalized_v = v_matrix / norm
        normalized_vectors_17a.append(normalized_v)

P_17a = Matrix.hstack(*normalized_vectors_17a)
print("Матрица поворота P для а):", P_17a)

# Диагональная матрица D
D_17a = P_17a.T * A_17a * P_17a
print("Диагональная матрица D для а):", D_17a)

# Диагональные элементы (числа)
D_vals_17a = [float(l.evalf()) for l in lambdas_a]  # Используем собственные значения напрямую
print("Диагональные коэффициенты для а):", D_vals_17a)

# Линейные коэффициенты в новых координатах
linear_new_17a = P_17a.T * b_vec_a
print("Линейные коэффициенты в новых координатах для а):", linear_new_17a)

# Центр (решение системы производных)
x0, y0 = symbols('x0 y0')
eq_full_a = a_a * x0**2 + b_a * x0 * y0 + c_a * y0**2 + b_vec_a[0] * x0 + b_vec_a[1] * y0 + f_a
partial_x_a = eq_full_a.diff(x0)
partial_y_a = eq_full_a.diff(y0)
sol_center_a = solve([partial_x_a, partial_y_a], [x0, y0])
print("Решение для центра а):", sol_center_a)
h_a = float(sol_center_a[x0].evalf())
k_a = float(sol_center_a[y0].evalf())
print("Центр кривой а) (h, k):", (h_a, k_a))

# Константа для канонического уравнения
# Правый член = -f - sum (d_i^2 / (4 * lambda_i))
d1_a = float(linear_new_17a[0].evalf())
d2_a = float(linear_new_17a[1].evalf())
lambda1_a, lambda2_a = D_vals_17a
right_side_a = -f_a - (d1_a**2 / (4 * lambda1_a) + d2_a**2 / (4 * lambda2_a))
print("Правая часть канонического уравнения для а):", right_side_a)

# Каноническое уравнение (с учетом порядка собственных значений)
if lambda1_a < lambda2_a:
    canonical_eq_a = f"{lambda1_a} \\xi^2 + {lambda2_a} \\eta^2 = {right_side_a}"
else:
    canonical_eq_a = f"{lambda2_a} \\xi^2 + {lambda1_a} \\eta^2 = {right_side_a}"
print("Каноническое уравнение а):", canonical_eq_a)

# Каноническая система координат: сдвиг к центру (h,k), затем поворот на угол, определяемый P
print("Каноническая система координат а): Сдвиг к центру (2, 3), поворот матрицей P = ", P_17a)

# б) 5x^2 + 12xy -22x -12y -19 = 0
a_b, b_b, c_b = 5, 12, 0
type_b = conic_type(a_b, b_b, c_b)
print("Тип кривой б):", type_b)

A_17b = Matrix([[5, 6], [6, 0]])  # Симметричная матрица (b/2)
b_vec_b = Matrix([-22, -12])  # Линейные коэффициенты
f_b = -19  # Константа

# Собственные значения
eigenvals_17b = A_17b.eigenvects(simplify=True)
lambdas_b = [val for val, _, _ in eigenvals_17b]
print("Собственные значения для б):", lambdas_b)

# Нормализованные собственные векторы
normalized_vectors_17b = []
for _, _, eigenvectors in eigenvals_17b:
    for v in eigenvectors:
        v_matrix = Matrix(v)
        norm = sqrt(v_matrix.dot(v_matrix))
        normalized_v = v_matrix / norm
        normalized_vectors_17b.append(normalized_v)

P_17b = Matrix.hstack(*normalized_vectors_17b)
print("Матрица поворота P для б):", P_17b)

# Диагональная матрица D
D_17b = P_17b.T * A_17b * P_17b
print("Диагональная матрица D для б):", D_17b)

# Диагональные элементы (числа)
D_vals_17b = [float(l.evalf()) for l in lambdas_b]
print("Диагональные коэффициенты для б):", D_vals_17b)

# Линейные коэффициенты в новых координатах
linear_new_17b = P_17b.T * b_vec_b
print("Линейные коэффициенты в новых координатах для б):", linear_new_17b)

# Центр
eq_full_b = a_b * x0**2 + b_b * x0 * y0 + c_b * y0**2 + b_vec_b[0] * x0 + b_vec_b[1] * y0 + f_b
partial_x_b = eq_full_b.diff(x0)
partial_y_b = eq_full_b.diff(y0)
sol_center_b = solve([partial_x_b, partial_y_b], [x0, y0])
print("Решение для центра б):", sol_center_b)
h_b = float(sol_center_b[x0].evalf())
k_b = float(sol_center_b[y0].evalf())
print("Центр кривой б) (h, k):", (h_b, k_b))

# Константа для канонического уравнения
d1_b = float(linear_new_17b[0].evalf())
d2_b = float(linear_new_17b[1].evalf())
lambda1_b, lambda2_b = D_vals_17b[0], D_vals_17b[1]  # -4, 9
right_side_b = -f_b - (d1_b**2 / (4 * lambda1_b) + d2_b**2 / (4 * lambda2_b))
print("Правая часть канонического уравнения для б):", right_side_b)

# Каноническое уравнение (с учетом знаков для гиперболы)
if lambda1_b < 0:
    canonical_eq_b = f"{lambda2_b} \\eta^2 + {lambda1_b} \\xi^2 = {right_side_b}"  # Положительный для v^2, отрицательный для u^2
else:
    canonical_eq_b = f"{lambda1_b} \\xi^2 + {lambda2_b} \\eta^2 = {right_side_b}"
print("Каноническое уравнение б):", canonical_eq_b)

# Каноническая система координат
print("Каноническая система координат б): Сдвиг к центру (1, 1), поворот матрицей P = ", P_17b)
```

## 1.18. При каких значениях параметра λ квадратичная форма Q(x) является положительно определенной, если

$ Q(x_1, x_2, x_3) =x_1^2 + 4x_2^2 + x_3^2 + 2λx_1x_2 + 10x_1x_3 + 6x_2x_3.$

```python
from sympy import Matrix, symbols, solve, lambdify
import numpy as np

# Определяем символическую переменную lambda
lam = symbols('lambda')

# Матрица квадратичной формы (коэффициенты 2*lam, 10, 6 делятся на 2 для a_ij)
A = Matrix([[1, lam, 5], [lam, 4, 3], [5, 3, 1]])

# Вычисляем миноры
det1 = A[0, 0]
det2 = A[0:2, 0:2].det()
det3 = A.det()

print("Условия для положительной определенности:")
print(f"det1 = {det1} > 0 (всегда истинно)")
print(f"det2 = {det2} > 0")
print(f"det3 = {det3} > 0")

# Решаем условие det2 > 0
det2_cond = solve(det2 > 0, lam)
print(f"det2 > 0: {det2_cond}")  # Ожидается -2 < lambda < 2

# Решаем условие det3 > 0
det3_cond = solve(det3 > 0, lam)
print(f"det3 > 0: {det3_cond}")

# Поскольку аналитическое решение сложное, проверяем собственные значения численно
lam_values = [-1.5, 0, 1.5]  # Тестовые значения в интервале (-2, 2)
for l_val in lam_values:
    A_sub = A.subs(lam, l_val)
    eigs = A_sub.eigenvals()  # Получаем словарь собственных значений
    # Преобразуем в список вещественных частей, если комплексные
    eigs_val = []
    for v in eigs:
        val = v.evalf()
        if val.is_real:  # Проверяем, является ли число вещественным
            eigs_val.append(float(val))
        else:  # Берем вещественную часть комплексного числа
            eigs_val.append(float(val.as_real_imag()[0]))
    print(f"lambda = {l_val}, собственные значения (вещественные части): {eigs_val}")
    if all(e > 0 for e in eigs_val):  # Проверяем положительность
        print(f"Положительно определена для lambda = {l_val}")
    else:
        print(f"Не положительно определена для lambda = {l_val}")

# Вывод: если ни одно значение не удовлетворяет, форма не положительно определена
print("Вывод: Нет значений lambda в проверяемом диапазоне (-2, 2), при которых Q(x) положительно определена.")
```

## 1.19. При каких значениях параметра λ квадратичная форма Q(x) является отрицательно определенной, если

 $ Q(x_1, x_2, x_3) =λx_1^2 −2x_2^2 −3x_3^2 + 2x_1x_2 −2x_1x_3 + 2x_2x_3.$

```python
from sympy import Matrix, symbols, solve

# Определяем символическую переменную lambda
lam = symbols('lambda')

# Матрица квадратичной формы
A = Matrix([[lam, 1, -1], [1, -2, 1], [-1, 1, -3]])

# Вычисляем миноры
det1 = A[0, 0]
det2 = A[0:2, 0:2].det()
det3 = A.det()

print("Условия для отрицательной определенности:")
print(f"det1 = {det1} < 0 => lambda < 0")
print(f"det2 = {det2} > 0 => lambda < -1/2")
print(f"det3 = {det3} < 0 => lambda < -1")

# Общий интервал: самое строгое - lambda < -1
print("Вывод: Q(x) отрицательно определена при lambda < -1.")

# Проверка на примере
lam_val = -2
A_test = A.subs(lam, lam_val)
eigs = A_test.eigenvals()
eigs_val = [float(v.evalf()) for v in eigs]
print(f"lambda = {lam_val}, собственные значения: {eigs_val}")
if all(e < 0 for e in eigs_val):
    print("Подтверждено: отрицательно определена")
```

## 1.20. Эквивалентны ли над полем вещественных чисел квадратичные формы Q_1 и Q_2, где

$ Q_1= 2x_1^2 + 9x_2^2 + 3x_3^2 + 8x_1x_2 −4x_1x_3 −10x_2x_3;$
$ Q_2= 2x_1 + 3x_2^2 + 6x_3^2 −4x_1x_2 −4x_1x_3 + 8x_2x_3?$

```python
from sympy import Matrix

# Матрицы квадратичных форм
A1 = Matrix([[2,4,-2],[4,9,-5],[-2,-5,3]])
A2 = Matrix([[2,-2,-2],[-2,3,4],[-2,4,6]])

# Собственные значения
eigs1 = A1.eigenvals()
eigs2 = A2.eigenvals()

eigs_val1 = [float(v.evalf()) for v in eigs1]
eigs_val2 = [float(v.evalf()) for v in eigs2]

print("Собственные значения Q1:", eigs_val1)
print("Собственные значения Q2:", eigs_val2)

# Количество положительных собственные значений
num_pos1 = sum(1 for e in eigs_val1 if e > 0)
num_neg1 = sum(1 for e in eigs_val1 if e < 0)
num_zero1 = sum(1 for e in eigs_val1 if e == 0)

num_pos2 = sum(1 for e in eigs_val2 if e > 0)
num_neg2 = sum(1 for e in eigs_val2 if e < 0)
num_zero2 = sum(1 for e in eigs_val2 if e == 0)

print(f"Q1: {num_pos1} положительных, {num_neg1} отрицательных, {num_zero1} нулевых")
print(f"Q2: {num_pos2} положительных, {num_neg2} отрицательных, {num_zero2} нулевых")

if num_pos1 == num_pos2 and num_neg1 == num_neg2 and num_zero1 == num_zero2:
    print("Q1 и Q2 эквивалентны над полем вещественных чисел, так как имеют одинаковую сигнатуру.")
else:
    print("Q1 и Q2 не эквивалентны.")
```
