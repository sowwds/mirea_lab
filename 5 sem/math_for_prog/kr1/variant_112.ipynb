## Задание 1

Матрица $A = (a_{i,j})$ имеет размер $n \times n$, где $n = 50000$. Известно, что ее диагональные элементы образуют арифметическую прогрессию с начальным элементом $a_{1,1} = 0.99$ и конечным элементом $a_{n,n} = 1.01$. Также известно, что все элементы матрицы $A$, расположенные вне главной диагонали, равны нулю за исключением двух элементов: $a_{1,2} = 0.59$ и $a_{2,1} = 0.13$. Найдите: 1) определитель матрицы $A$; 2) след матрицы, обратной к $A$.

```python
import numpy as np

n = 50000
a11 = 0.99
ann = 1.01
a12 = 0.59
a21 = 0.13

d = np.linspace(a11, ann, n)

det_A = (d[0] * d[1] - a12 * a21) * np.prod(d[2:])
print(f"1) Определитель матрицы А: {det_A}")

inv_det_block = 1 / (d[0] * d[1] - a12 * a21)
b11 = d[1] * inv_det_block
b22 = d[0] * inv_det_block

trace_B = b11 + b22 + np.sum(1 / d[2:])
print(f"2) След матрицы, обратной к A: {trace_B}")
```

## Задание 2

Даны матрицы: $I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$, $A = \begin{pmatrix} 9\gamma & 11\gamma \\ 6\gamma & 3\gamma \end{pmatrix}$, где $\gamma = 0.059$. Пусть $S_{m,n} = \sum_{i=m}^{n-1} A^i$, $B_m = (I - A)^{-1} A^m$. 1) Найдите наибольший элемент матрицы $S = S_{4,90}$. 2) Найдите наибольший элемент матрицы $B = B_4$. 3) Найдите наименьшее $k$, такое, что при любом $n > k$ наибольший элемент матрицы $S_{k,n}$ меньше 0,0002.

```python
import numpy as np
from numpy.linalg import inv, matrix_power

gamma = 0.059
A = gamma * np.array([[9, 11], [6, 3]])
I = np.identity(2)

S_4_90 = np.zeros_like(A)
for i in range(4, 90):
    S_4_90 += matrix_power(A, i)
max_S = np.max(S_4_90)
print(f"1) Наибольший элемент матрицы S = S4,90: {max_S}")

B4 = inv(I - A) @ matrix_power(A, 4)
max_B4 = np.max(B4)
print(f"2) Наибольший элемент матрицы B = B4: {max_B4}")

eigenvalues, _ = np.linalg.eig(A)
max_eigenvalue = np.max(np.abs(eigenvalues))

C = np.linalg.inv(I - A)
k = 1
while True:
    Sk_inf = np.dot(matrix_power(A, k), C)
    if np.max(np.abs(Sk_inf)) < 0.0002:
        print(f"3) Наименьшее k: {k}")
        break
    k += 1
```

## Задание 5

Дана матрица: $A = \begin{pmatrix} 14 & 21 & 24 \\ 10 & 17 & 29 \\ 23 & 13 & 20 \end{pmatrix}$. Пусть $Z$ – собственное значение матрицы $A$ с наибольшей мнимой частью. Найдите $Z$ и соответствующий этому собственному значению собственный вектор $X$, первая координата которого равна 1, $X = (1, U, V)$. В ответе укажите: 1) мнимую часть $Z$; 2) действительную часть $U$.

```python
import numpy as np

A = np.array([[14, 21, 24], [10, 17, 29], [23, 13, 20]])
eigenvalues, eigenvectors = np.linalg.eig(A)

Z = eigenvalues[np.argmax(np.imag(eigenvalues))]
X_unnormalized = eigenvectors[:, np.argmax(np.imag(eigenvalues))]

X = X_unnormalized / X_unnormalized[0]
U = X[1]

print(f"1) Мнимая часть Z: {np.imag(Z)}")
print(f"2) Действительная часть U: {np.real(U)}")
```

## Задание 6

Для каждого $k = 1, \dots, n$ обозначим $e_k$ вектор-столбец размерности $n$, такой, что его элемент с номером $k$ равен 1, a все остальные элементы равны 0. Пусть $\sigma: \{1, \dots, n\} \to \{1, \dots, n\}$ – некоторая перестановка первых $n$ натуральных чисел. Матрица $A$ однозначно определяется условиями: $Ae_k = e_{\sigma(k)}$, $k = 1, \dots, n$. Известно, что $n = 12$ и результатом применения $\sigma$ к последовательности $1, \dots, 12$ будет следующий ряд чисел: 2, 1, 6, 4, 10, 11, 12, 8, 7, 3, 9, 5. Найдите: 1) след матрицы $A^4$; 2) наименьшее натуральное $m$, такое, что след матрицы $A^m$ равен 12.

```python
import numpy as np
from math import gcd

def lcm(a, b):
    return abs(a*b) // gcd(a, b) if a != 0 and b != 0 else 0

sigma_map = {1: 2, 2: 1, 3: 6, 4: 4, 5: 10, 6: 11, 7: 12, 8: 8, 9: 7, 10: 3, 11: 9, 12: 5}
n = 12
visited = [False] * (n + 1)
cycles = []
for i in range(1, n + 1):
    if not visited[i]:
        cycle = []
        j = i
        while not visited[j]:
            visited[j] = True
            cycle.append(j)
            j = sigma_map[j]
        cycles.append(cycle)

cycle_lengths = [len(c) for c in cycles]

trace_A4 = 0
for length in cycle_lengths:
    if 4 % length == 0:
        trace_A4 += length
print(f"1) След матрицы А^4: {trace_A4}")

m = 1
for length in cycle_lengths:
    m = lcm(m, length)
print(f"2) Наименьшее натуральное m: {m}")
```
